### 这是一个 Notebook

本地编辑，本地运行。

markdown block 不会显示运行需要，但是 code block 会显示。

有些代码仅仅为了生成图像，代码应该隐藏。

而且 notebook 不容易内嵌元数据，无法仅从一个 notebook 分析出创作时间、分类、标签。

因此，还是要使用 markdown，最多把 notebook 嵌入 markdown 内部。

Notebook 里面，块可以折叠。展开的就显示出来，折叠的就隐藏。这样就实现了绘图代码的隐藏。

```python
import os
import pandoc
from pandoc.types import *
```

```python
with open('sample_markdown.md', 'r') as f:
    txt = f.read()
```

```python
# 解析文档
doc = pandoc.read(txt)
```

```python
# 文档就是AST，是一棵树

# 使用iter可以遍历一棵树，先序遍历
for blk,loc in pandoc.iter(doc, path=True):
    print('content of block:', loc)
```
